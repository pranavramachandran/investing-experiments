```python
import os.path
import pandas as pd

def isDownloaded(date):
    fileName = "../data/" + date.strftime('%Y-%m-%d')
    return os.path.exists(fileName)

def readData(date):
    fileName = "../data/" + date.strftime('%Y-%m-%d')
    df = pd.read_csv(fileName)
    return df

# Iterate through dates
import datetime
today = datetime.date.today()
twoYearsAgo = today - datetime.timedelta(days=365*2)
start_date = today - datetime.timedelta(days=30)
end_date = today #datetime.date(2022, 1, 31)
delta = datetime.timedelta(days=1)

df = pd.DataFrame()

current_date = start_date
while current_date <= end_date:
    if current_date.weekday() < 5:
        print("Processing " + current_date.strftime('%Y-%m-%d') + "... ",end='')
        if isDownloaded(current_date):
            df1 = readData(current_date)
            print("Read " + str(len(df1)) + " rows.")
            frames = [df, df1]
            df = pd.concat(frames)
        else:
            print("not downloaded.")
    else:
        print("Skipping " + current_date.strftime('%Y-%m-%d'))
    current_date += delta

df = df.reset_index(drop=True)
```
